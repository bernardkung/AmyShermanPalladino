#### Overall Goals

Data Sources:
- Gilmore Girls: scripts are typical screenplays, and it's a matter of counting words and running these counts against the runtimes scraped out of imdb
- Marvelous Mrs. Maisel: scripts are transcripts: dialog is identified every time a new character speaks, but otherwise not well attributed.
Need to strip out names, actions/directions, and songs, and then combine with runtimes.
- Bunheads: no scripts or transcripts yet. So that's unfortunate.

```python
import json
```

```python
def open_json(file_path):
    with open(file_path, 'r') as f:
        return json.load(f)
```

```python
gilmore_scripts = open_json('./output/gilmore_scripts.json')
gilmore_runtimes = open_json('./output/gilmore_runtimes.json')
```

```python
curr_season = 1
print("", 1, end = "\n", flush=True)
for script in gilmore_scripts:
    if script['season_number'] == curr_season:
        print(script['episode_number'], end =" ", flush=True)
    elif script['season_number'] != curr_season:
        curr_season = script['season_number']
        print("\n", script['season_number'], end = "\n", flush=True)
```
